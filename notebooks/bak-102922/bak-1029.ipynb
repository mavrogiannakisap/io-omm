```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
nchdd = pd.read_csv('hdd_nocache_res.csv')
chdd = pd.read_csv('hdd_cache_res.csv')
ncssd = pd.read_csv('ssd_nocache_res.csv')
cssd = pd.read_csv('ssd_cache_res.csv')
allb = [nchdd, chdd, ncssd, cssd]
labels = ['NoCache HDD', 'Cached HDD',
          'NoCache SSD', 'Cached SSD']
```

```python
def dfselect(df, col, val, drop=False):
    res = df[df[col] == val]
    return res.drop([col], axis=1) if drop else res
for x in allb:
    x['nlg'] = np.log2(x['n'])
```

```python
fig, axs = plt.subplots(4, sharex=False, sharey=True, figsize=(8, 14))
plt.subplots_adjust(hspace=.45)

allb = [nchdd, chdd, ncssd, cssd]
labels = ['NoCache HDD', 'NoCache SSD', 'Cached HDD', 'Cached SSD']
for idx, d in enumerate([allb[0], allb[2], allb[1], allb[3]]):
    ax = axs[idx]
    
    po = dfselect(d, 'test', 'poram')
    bo = dfselect(d, 'test', 'boram')
    
    ax.plot(po['c'], po['itime'], label='Path ORAM', color='r')
    ax.plot(bo['c'], bo['itime'], label='Block ORAM', color='b')
    
    ax.set_title(f'{labels[idx]}')
    ax.set_xlabel('Count')
    ax.set_ylabel('Insert Time (s)')
    ax.set_xscale('log', base=2)
    ax.set_yscale('log')
    ax.grid()
    ax.legend(loc='upper left')

plt.show()
```

```python
allb = [cssd, ncssd, nchdd]
labels = ['SSD', 'SSD (No Cache)', 'HDD (No Cache)']
for idx, d in enumerate(allb):
    fig, axs = plt.subplots(1, sharex=False, sharey=True, figsize=(8, 3.5))
    plt.subplots_adjust(hspace=.45)
    ax = axs
#     ax = axs[idx]
    
    po = dfselect(d, 'test', 'poram')
    bo = dfselect(d, 'test', 'boram')
    
    ax.plot(po['c'], po['itime'], label='Path ORAM', color='r')
    ax.plot(bo['c'], bo['itime'], label='Block ORAM', color='b')
    
#     ax.set_title(f'{labels[idx]}')
    ax.set_title('Access Time')
    ax.set_xlabel('Value Size (#blocks)')
    ax.set_ylabel('Insert Time (s)')
    ax.set_xscale('log', base=2)
    ax.set_yscale('log')
    ax.grid()
    ax.legend(loc='upper left')
    plt.savefig(f'{idx+1}-bencher.eps', bbox_inches='tight')

plt.show()
```
