```python
nchdd = pd.read_csv('hdd_nocache_res.csv')
chdd = pd.read_csv('hdd_cache_res.csv')
ncssd = pd.read_csv('ssd_nocache_res.csv')
cssd = pd.read_csv('ssd_cache_res.csv')
nnvme = pd.read_csv('nvme_nocache_res.csv')
cnvme = pd.read_csv('nvme_cache_res.csv')
allb = [chdd, cssd, cnvme, nchdd, ncssd, nnvme]
labels = ['Cached HDD', 'Cached SSD', 'Cached NVMe SSD',
          'NoCache HDD', 'NoCache SSD', 'NoCache NVMe SSD']
```

```python
chdd
```

```python
def dfselect(df, col, val, drop=False):
    res = df[df[col] == val]
    return res.drop([col], axis=1) if drop else res
```

```python
for x in allb:
    x['nlg'] = np.log2(x['n'])
```

```python
fig, axs = plt.subplots(6, sharex=False, sharey=True, figsize=(8, 21))
plt.subplots_adjust(hspace=.45)

for idx, d in enumerate(allb):
    ax = axs[idx]
    
    po = dfselect(d, 'test', 'poram')
    bo = dfselect(d, 'test', 'boram')
    
    ax.plot(po['c'], po['itime'], label='Path ORAM', color='r')
    ax.plot(bo['c'], bo['itime'], label='Block ORAM', color='b')
    
    ax.set_title(f'{labels[idx]}')
    ax.set_xlabel('Count')
    ax.set_ylabel('Insert Time (s)')
    ax.set_xscale('log', base=2)
    ax.set_yscale('log')
    ax.grid()
    ax.legend(loc='upper left')

plt.show()
```

```python
fig, axs = plt.subplots(2, sharex=False, sharey=True, figsize=(8, 7))
plt.subplots_adjust(hspace=.45)

for idx, d in enumerate(allb[:2]):
    ax = axs[idx]
    
    po = dfselect(d, 'test', 'poram')
    bo = dfselect(d, 'test', 'boram')
    
    ax.plot(po['c'], po['itime'], label='Path ORAM', color='r')
    ax.plot(bo['c'], bo['itime'], label='Block ORAM', color='b')
    
    ax.set_title(f"{['HDD', 'SSD'][idx]}")
    ax.set_xlabel('Count')
    ax.set_ylabel('Insert Time (s)')
    ax.set_xscale('log', base=2)
    ax.set_yscale('log')
    ax.grid()
    ax.legend(loc='upper left')

plt.savefig('bencher.eps')
plt.show()
```
